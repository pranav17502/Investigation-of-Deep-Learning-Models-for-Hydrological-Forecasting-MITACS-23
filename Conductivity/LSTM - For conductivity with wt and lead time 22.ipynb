We pass different types of parameters to the model, like feautures, target variable and now we also want our model to use the windowing. So the parameters of windowing like the sampling rate, the length of a window, the stride and all we have to pass to the model. This is done by the object of TimeseriesGenerator. When we see the configuration of it, we see a dictionary. The parameters which we want to pass to the model are the "keys" of the dictionary and the parameter values which we actually pass to the model is the "value" of that particular parameter (i.e. "key"). We can see that above.

# **Building Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

##MAE

```python
model_mae = Sequential()
model_mae.add(LSTM(25, activation='relu', return_sequences=True, input_shape=(24, 45)))
model_mae.add(LSTM(25, activation='relu', return_sequences=True))
model_mae.add(LSTM(25, activation='relu'))
model_mae.add(Dense(1))
```

```python
from keras.callbacks import ModelCheckpoint
```

```python
checkpoint = ModelCheckpoint('best_weights.h5', monitor='val_loss', save_best_only=True)
```

```python
model_mae.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),loss='mae', metrics= ["mse"])
model_mae.summary()
```

```python
history = model_mae.fit(train_generator, validation_data= test_generator, epochs=100, callbacks=[checkpoint])
```

```python
model_mae.load_weights('best_weights.h5')
model_mae.save_weights('best_weights.h5')
```

```python
plt.plot(history.history['mse'])
plt.plot(history.history['val_mse'])
plt.title('model mse')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
