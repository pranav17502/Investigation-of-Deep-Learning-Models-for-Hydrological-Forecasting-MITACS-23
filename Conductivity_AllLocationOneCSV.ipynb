# **Interpolation of Missing values**

```python
conductivity = pd.read_csv("Conductivity.csv")
conductivity
```

```python
conductivity.set_index("Timestamp", inplace= True)
```

```python
conductivity
```

```python
#Percentage of Missing values
l = len(conductivity)
for i in range(1, len(conductivity.columns)):
  m = conductivity.iloc[ : , i].isin([-999]).sum(axis= 0)
  p = 100*m/l
  print(conductivity.columns[i], "         ",p)
```

```python
conductivity[conductivity == -999] = np.nan
conductivity.isin([np.nan]).sum(axis= 0)
```

```python
#Percentage of Missing values
l = len(conductivity)
for i in range(1, len(conductivity.columns)):
  m = conductivity.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(conductivity.columns[i], "         ",p)
```

```python
conductivity_linear  = conductivity.interpolate(method= "linear")
```

```python
conductivity_linear
```

```python
#Percentage of Missing values
l = len(conductivity)
for i in range(1, len(conductivity.columns)):
  m = conductivity_linear.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(conductivity_linear.columns[i], "         ", p)
```

```python
conductivity.isna().sum()
```

```python
#Checking for negative values as decomposition does not work with negative and zero values
conductivity_linear[conductivity_linear < 0].sum()
```

```python
conductivity_linear[conductivity_linear < 0].count()
```

```python
conductivity.interpolate(method= "linear", inplace= True)
```

```python
conductivity[conductivity < 0].count()
```

There are no negative values now.
