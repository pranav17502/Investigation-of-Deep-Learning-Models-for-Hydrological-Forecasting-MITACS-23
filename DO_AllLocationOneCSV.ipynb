# Combining all location data

```python
main_df = pd.DataFrame()
```

```python
t = pd.Series(pd.date_range(start= '2018-01-01 00:00:00', end= '2023-01-01 23:00:00', freq= 'H'))
main_df.insert(0, 'Timestamp', t)
main_df
```

# **Interpolation of Missing values**

```python
DO = pd.read_csv("DO.csv")
DO
```

```python
DO.set_index("Timestamp", inplace= True)
```

```python
DO
```

```python
#Percentage of Missing values
l = len(DO)
for i in range(1, len(DO.columns)):
  m = DO.iloc[ : , i].isin([-999]).sum(axis= 0)
  p = 100*m/l
  print(DO.columns[i], "         ",p)
```

```python
DO[DO == -999] = np.nan
DO.isin([np.nan]).sum(axis= 0)
```

```python
#Percentage of Missing values
l = len(DO)
for i in range(1, len(DO.columns)):
  m = DO.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(DO.columns[i], "         ",p)
```

```python
DO_linear  = DO.interpolate(method= "linear")
```

```python
DO_linear
```

```python
#Percentage of Missing values
l = len(DO)
for i in range(1, len(DO.columns)):
  m = DO_linear.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(DO_linear.columns[i], "         ", p)
```

```python
#Checking for negative values as decomposition does not work with negative and zero values
DO_linear[DO_linear < 0].sum()
```

```python
DO_linear[DO_linear < 0].count()
```

```python
DO.interpolate(method= "linear", inplace= True)
```

```python
DO[DO < 0].count()
```

There are no negative values now.
