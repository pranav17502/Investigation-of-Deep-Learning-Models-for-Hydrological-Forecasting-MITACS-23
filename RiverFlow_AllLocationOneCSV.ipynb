```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
main_df = pd.DataFrame()
```

```python
t = pd.Series(pd.date_range(start= '2018-01-01 00:00:00', end= '2023-01-01 23:00:00', freq= 'H'))
main_df.insert(0, 'Timestamp', t)
main_df
```

Aberfoyle

```python
df = pd.read_csv('Aberfoyle.csv')
df.head()
```

```python
df.Timestamp = pd.to_datetime(df.Timestamp)
df.head()
```

```python
main_df = main_df.merge(df[["Timestamp","Value"]], on="Timestamp", how="left")
main_df["Value"] = main_df["Value"].fillna(-999)
main_df
```

```python
main_df = main_df.rename(columns= {'Value': 'Aberfoyle'})
main_df
```
