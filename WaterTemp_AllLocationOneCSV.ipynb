# **Interpolation of Missing values**

```python
wt = pd.read_csv("Water_Temperature.csv")
wt
```

```python
wt.set_index("Timestamp", inplace= True)
```

```python
wt
```

```python
#Percentage of Missing values
l = len(wt)
for i in range(1, len(wt.columns)):
  m = wt.iloc[ : , i].isin([-999]).sum(axis= 0)
  p = 100*m/l
  print(wt.columns[i], "         ",p)
```

Dropping Mill Creek, New Dundee, Mount Vernon and Below Shand

```python
wt.drop(columns= ["Mill Creek", "New Dundee", "Mount Vernon", "Below Shand"], inplace= True)
wt
```

```python
wt[wt == -999] = np.nan
wt.isin([np.nan]).sum(axis= 0)
```

```python
#Percentage of Missing values
l = len(wt)
for i in range(1, len(wt.columns)):
  m = wt.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(wt.columns[i], "         ",p)
```

```python
wt_linear  = wt.interpolate(method= "linear")
```

# For Aberfoyle

```python
aberfoyle = wt["Aberfoyle"]
```

```python
aberfoyle.index = pd.to_datetime(aberfoyle.index)
```

```python
aberfoyle.isna().sum()
```

```python
aberfoyle.index.dtype
```

```python
from dateutil.relativedelta import relativedelta
```

```python
for i in range(0, 1192):
  dt = aberfoyle.index[i]
  dt1 = dt + relativedelta(years=1)
  dt2 = dt + relativedelta(years=2)
  dt3 = dt + relativedelta(years=3)
  dt4 = dt + relativedelta(years=4)
  aberfoyle[aberfoyle.index == dt] = (aberfoyle[aberfoyle.index == dt1].values + aberfoyle[aberfoyle.index == dt2].values + aberfoyle[aberfoyle.index == dt3].values + aberfoyle[aberfoyle.index == dt4].values)/4
```

```python
aberfoyle.isna().sum()
```

```python
wt["Aberfoyle"] = aberfoyle
```

```python
wt.isna().sum()
```
