# **Interpolation of Missing values**

```python
pH = pd.read_csv("pH.csv")
pH
```

```python
pH.set_index("Timestamp", inplace= True)
```

```python
pH
```

```python
#Percentage of Missing values
l = len(pH)
for i in range(1, len(pH.columns)):
  m = pH.iloc[ : , i].isin([-999]).sum(axis= 0)
  p = 100*m/l
  print(pH.columns[i], "         ",p)
```

```python
pH[pH == -999] = np.nan
pH.isin([np.nan]).sum(axis= 0)
```

```python
#Percentage of Missing values
l = len(pH)
for i in range(1, len(pH.columns)):
  m = pH.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(pH.columns[i], "         ",p)
```

```python
pH_linear  = pH.interpolate(method= "linear")
```

```python
pH_linear
```

```python
#Percentage of Missing values
l = len(pH)
for i in range(1, len(pH.columns)):
  m = pH_linear.iloc[ : , i].isin([np.nan]).sum(axis= 0)
  p = 100*m/l
  print(pH_linear.columns[i], "         ", p)
```

```python
#Checking for negative values as decomposition does not work with negative and zero values
pH_linear[pH_linear < 0].sum()
```

```python
pH_linear[pH_linear < 0].count()
```

```python
pH.interpolate(method= "linear", inplace= True)
```

```python
pH[pH < 0].count()
```

There are no negative values now.

# Interpolating the 95 missing values of Victoria

```python
victoria = pH["Breslau Victoria"]
victoria
```

```python
victoria.isna().sum()
```

```python
victoria.index = pd.to_datetime(victoria.index)
```

```python
victoria[94], victoria[95]
```

```python
from dateutil.relativedelta import relativedelta
```

```python
for i in range(0, 95):
  dt = victoria.index[i]
  dt1 = dt + relativedelta(years=1)
  dt2 = dt + relativedelta(years=2)
  dt3 = dt + relativedelta(years=3)
  dt4 = dt + relativedelta(years=4)
  victoria[victoria.index == dt] = (victoria[victoria.index == dt1].values + victoria[victoria.index == dt2].values + victoria[victoria.index == dt3].values + victoria[victoria.index == dt4].values)/4
```

```python
victoria.isna().sum()
```

```python
pH.index = pd.to_datetime(pH.index)
```

```python
pH["Breslau Victoria"] = victoria
```

```python
pH.isna().sum()
```

```python
pH.to_csv("pH_noNA.csv")
files.download("pH_noNA.csv")
```
